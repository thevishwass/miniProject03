```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/2020-2025.csv')
```

```python
big_countries = df[df['Country'].isin(['India', 'China', 'United States'])]

plt.figure(figsize=(8,6))
sns.barplot(x='Country', y='2023', data=big_countries, palette='viridis')
plt.title('2023 GDP for India, China, USA', fontsize=16)
plt.ylabel('Value', fontsize=12)
plt.xlabel('Country', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Select India data
india_data = df[df['Country'] == 'India'].iloc[0, 1:]  # all years columns

plt.figure(figsize=(10,6))
india_data.plot(kind='line', marker='o', color='orange')
plt.title('India Data Trend: 2020-2025', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Value in Trillion USD', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
#  Filter dataset for major countries
big_countries = ['India', 'China', 'United States']
selected = df[df['Country'].isin(big_countries)].set_index('Country').T

# Reset the index to have years as a column for better plotting
selected = selected.reset_index().rename(columns={'index': 'Year'})

# Plot
plt.figure(figsize=(10,6))
sns.lineplot(data=selected, x='Year', y='India', marker='o', label='India', color='orange', linewidth=2)
sns.lineplot(data=selected, x='Year', y='China', marker='s', label='China', color='green', linewidth=2)
sns.lineplot(data=selected, x='Year', y='United States', marker='^', label='United States', color='blue', linewidth=2)

# Titles and labels
plt.title('Trends from 2020 to 2025 for Major Countries', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Country', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Filter data
big_countries = df[df['Country'].isin(['India', 'China', 'United States'])].set_index('Country')

# Prepare data for radar chart
labels = big_countries.columns
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # complete the loop

plt.figure(figsize=(8,8))
for country in big_countries.index:
    values = big_countries.loc[country].tolist()
    values += values[:1]  # close the loop
    plt.polar(angles, values, marker='o', label=country, linewidth=2)

plt.xticks(angles[:-1], labels)
plt.title("Radar Chart: 2020-2025 Values for Big Countries", fontsize=16)
plt.legend(loc='upper right')
plt.show()
```
