```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('../dataset/2020-2025.csv')  # make sure the path is correct!

# Now you can explore it:
df.info()
print(df.describe())
print(df.isnull().sum())
print(df.head())
print(df.columns)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Country', y='2023', data=df)
plt.title('2023 Values by Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Select one country, e.g., 'Albania'
india_data = df[df['Country'] == 'India'].iloc[0, 1:]  # selects all year columns

# Plotting
india_data.plot(kind='line', marker='o')
plt.title('India Data from 2020 to 2025')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(df['2022'].dropna(), shade=True)
plt.title('Distribution of 2022 Values Across Countries')
plt.xlabel('Value')
plt.tight_layout()
plt.show()
```
